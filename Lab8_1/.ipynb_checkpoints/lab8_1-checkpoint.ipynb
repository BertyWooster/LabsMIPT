## Изучение работы оптического пирометра 
Среднее значение яркостной температуры модели АЧТ $ 715^{\circ} C $. 

По хромель-алюмелевой термопаре $ 745^{\circ} C$.

Различие между значениями температуры, полученными разными способами, около $ 4 \% $. Oптический пирометр работает исправно.

## Проверка закона Стефана-Больцмана.

```python
#Проведение эксперимента:
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib import mlab
from scipy import interpolate
import math 
import warnings
warnings.filterwarnings('ignore')

def myFunction(x, a, b): 
     return a*x+b

# Парсинг данных.
data = pd.read_table('measurements.txt', sep = '\s+' )
Tbr = data['Tяр']# deg
Ttd = data['Ттерм']# deg 
I = data['I']# мА
V = data['V']# В

# Для каждого значения термодинамической темпиратуры вычисляю мощность, потребляемую нитью.
W = I*V # мВт

# Находим логарифм мощности:
lnW = W.copy()
for i in range(len(W)):
    lnW[i] = math.log(W[i]*10**(-3))
    
# Находим логарифм рмодинамической темпиратуры:
lnTtd = W.copy()
for i in range(len(Ttd)):
    lnTtd[i] = math.log(Ttd[i])

```

## Строю график зависимости потребляемой энергии от термодинамической темпиратуры.

```python
plt.figure(figsize = (10,7))
plt.plot(Ttd, W, 'go',color = 'brown')
plt.xlabel('Т, К')
plt.ylabel('W, мВт')
plt.title("W = W(T)")
plt.grid()
plt.legend();
```

## Теперь проводим построения в логарифмическом масштабе.

```python
def linear_fit(x, a, b):
    return a*x + b

result = curve_fit(f = linear_fit, xdata = lnTtd[4:10], ydata = lnW[4:10])
popt, pcov = result
popt

plt.figure(figsize = (10,7))
plt.plot(lnTtd, lnW, 'go',color = 'brown')
plt.plot(lnTtd, linear_fit(lnTtd, *popt))
plt.xlabel('lnT')
plt.ylabel('lnW')
plt.title("lnW(lnT)")
plt.grid()
plt.legend();
```

$ Из \ линейной \ аппроксимации \ высокотемпературного \ участка \ кривой \ получаем, \ что \ значение \\ коэффициента \ n \  в \ формуле \ \ln W = \ln (\varepsilon_{T} B) + n \ln T ,n \approx 3,9 . $
